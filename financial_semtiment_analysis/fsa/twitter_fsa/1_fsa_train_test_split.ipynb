```python
from datasets import load_dataset
```

```python
dataset = load_dataset("zeroshot/twitter-financial-news-sentiment")
dataset
```

```python
train_data = dataset["train"]
val_data = dataset["validation"]
print(f"Train dataset size: {len(train_data)}")
print(f"Validation dataset size: {len(val_data)}")
```

```python
train_data.set_format("pandas")
train_df = train_data[:].reset_index(drop=True)
train_df.head()
```

```python
val_data.set_format("pandas")
val_df = val_data[:]
len(val_df)
```

```python
train_df.to_csv("twitter_fsa/data/train.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split

val_data, test_data = train_test_split(val_df, test_size=0.2, stratify=val_df["label"], random_state=42, shuffle=True)
print(f"length of validation_data = {len(val_data)}")
print(f"length of test_data = {len(test_data)}")
```

```python
val_data.to_csv("twitter_fsa/data/validation.csv", index=False)
test_data.to_csv("twitter_fsa/data/test.csv", index=False)
```
